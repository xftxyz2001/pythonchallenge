```python
import zipfile

files = {}

fz = zipfile.ZipFile(r'c:\Users\25810\pythonchallenge\6_channel.zip')
for name in fz.namelist():
    with fz.open(name) as f:
        files[name] = f.read().decode('utf-8')
print(files['readme.txt'])
```
