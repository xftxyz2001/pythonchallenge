```python
a = ['1']
for i in range(30):
    m = []
    for c in a[-1]:
        if len(m) > 0 and m[-1][0] == c:
            m[-1][1] += 1
        else:
            m.append([c, 1])
    a.append(''.join(['{}{}'.format(n, c) for c, n in m]))
print(len(a[30]))
```
