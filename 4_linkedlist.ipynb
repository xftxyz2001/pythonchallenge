```python
import requests
url = 'http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=12345'
while True:
    r = requests.get(url)
    if not r.text.startswith('and'):
        print(r.text)
    if 'Divide' in r.text:
        num = int(url.split('=')[-1])/2
    else:
        num = r.text.split(' ')[-1]
    url = 'http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=' + \
        str(num)
```
