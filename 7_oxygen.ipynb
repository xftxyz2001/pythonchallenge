```python
from PIL import Image
img = Image.open(r"c:\Users\25810\pythonchallenge\7_oxygen.png")
w, h = img.size
pic = img.load()
h //= 2
lr, lg, lb, lx = 0, 0, 0, 0
rl, gl, bl, xl = [], [], [], []
for x in range(w):
    r, g, b, x = pic[x, h]
    if r != lr or g != lg or b != lb or x != lx:
        rl.append(r)
        gl.append(g)
        bl.append(b)
        xl.append(x)
    lr, lg, lb, lx = r, g, b, x
print(''.join(map(chr, rl)))
print(''.join(map(chr, gl)))
print(''.join(map(chr, bl)))
print(''.join(map(chr, xl)))
```
